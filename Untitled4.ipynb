```python
# Entrada

nota1 = float(input("digite sua nota do 1º bimestre "))

if nota1 < 0 or nota1 > 10 :
  print("nota invalida digite sua nota de 0 a 10:")
  nota1 = float(input("digite sua nota do 1º bimestre "))

nota2 = float(input("digite sua nota do 2º bimestre "))
if nota2 < 0 or nota2 > 10 :
  print("nota invalida digite sua nota de 0 a 10:")
  nota2 = float(input("digite sua nota do 2º bimestre "))

nota3 = float(input("digite sua nota do 3º bimestre "))
if nota3 < 0 or nota3 > 10 :
  print("nota invalida digite sua nota de 0 a 10:")
  nota3 = float(input("digite sua nota do 3º bimestre "))

nota4 = float(input("digite sua nota do 4º bimestre "))
if nota4 < 0 or nota4 > 10 :
  print("nota invalida digite sua nota de 0 a 10:")
  nota4 = float(input("digite sua nota do 4º bimestre "))

media = (nota1 + nota2 + nota3 + nota4)/4

# Processamento / saida


if media < 5 :
  print (f"Você foi Reprovado, sua media é : {media}")
elif media < 7 :
   print (f"Você está de recuperação, sua media é : {media}")
else:
 print (f"Você foi Aprovado, sua media é : {media}")
```

```python
#Escrever um programa que leia o valor da compra e aplique:

# Até R$ 100 - sem desconto
# R$ 101 - a R$ 500 - 10% de desconto
#Acima de R$ 500 - desconto de 20%

#Mostrar valor a pagar depois de aplicado o desconto.

compra = float(input("Digite o valor total da compra: R$ "))

if compra > 100:
    desconto = compra - (compra * 0.10)
    print (f"o valor a pagar é {desconto}")

elif compra > 500:
    desconto = compra - (compra * 0.20)
    print (f"o valor a pagar é: {desconto}")

else:
   print  (f"você não obteve desconto, valor a pagar : {compra}")
```
