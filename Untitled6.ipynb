```python
#1 Contagem simples o usuario digita um numero e o programa imprimi do zero  ate o numero

numero = int(input("Digite um número para iniciar a contagem: "))

print(f"Contando de 0 até {numero}:")
for i in range(numero + 1):
    print(i)
```

```python
# 2 - Contagem regrassiva, o usuario digita um número e o programa impprimi do número até o zero

numero = int(input("Digite um número para iniciar a contagem regressiva: "))

print(f"Contando regressivamente de {numero} até 0:")
for i in range(numero, -1, -1):
    print(i)
```

```python
# 3 - Números pares, o úsuario digita um número e o programa imprimi os numeros pares.

numero = int(input("Digite um número para encontrar os números pares: "))

print(f"Números pares de 0 até {numero}:")
for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

```python
# 4 - Números impares, o úsuario digita um número e o programa imprimi os numeros impares

numero = int(input("Digite um número para encontrar os números ímpares: "))

print(f"Números ímpares de 0 até {numero}:")
for i in range(numero + 1):
    if i % 2 != 0:
        print(i)
```

```python
# 5 - Soma, o usuario digita um número e o programa soma os numeros do zero até o número

numero = int(input("Digite um número para somar: "))

soma = 0
for i in range(numero + 1):
    soma += i

print(f"A soma dos números de 0 até {numero} é: {soma}")
```

```python
# 6 - Tabuada, o usuario digita um numero e o programa imprimi a tabuada so número.


numero = int(input("Digite um número para ver a tabuada: "))

print(f"Tabuada do {numero}:")
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

```python
# 7 - Contar letras, o usuario digita uma palavra e o programa conta as letras da palavras.

palavra = input("Digite uma palavra para contar as letras: ")

numero_letras = len(palavra)

print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
# 8 - Contar vogais, o usuario digita uma palavra e o programa conta as vogais.

palavra = input("Digite uma palavra para contar as vogais: ")
vogais = "aeiouAEIOU"
count_vogais = 0
for char in palavra:
    if char in vogais:
        count_vogais += 1
print(f"A palavra '{palavra}' tem {count_vogais} vogais.")
```

```python
# 9 - Maior número digitado, o usuario digita quatro números e o programa imprimi o maior número digitado.

print("Digite quatro números para encontrar o maior:")
numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))
numero3 = int(input("Digite o terceiro número: "))
numero4 = int(input("Digite o quarto número: "))

maior_numero = numero1

if numero2 > maior_numero:
    maior_numero = numero2
if numero3 > maior_numero:
    maior_numero = numero3
if numero4 > maior_numero:
    maior_numero = numero4

print(f"O maior número digitado é: {maior_numero}")
```

```python
#10 o usuario digita um numero entre 01 e 100 e o programa mostra quais são divisiveis por 3 e o programa mostra quais são divisiveis por 3 e 5

numero = int(input("Digite um número entre 1 e 100: "))

if 1 <= numero <= 100:
    print(f"Números divisíveis por 3 até {numero}:")
    for i in range(1, numero + 1):
        if i % 3 == 0:
            print(i)

    print(f"\nNúmeros divisíveis por 3 e 5 até {numero}:")
    for i in range(1, numero + 1):
        if i % 3 == 0 and i % 5 == 0:
            print(i)
elif numero < 1:
    print("Por favor, digite um número maior ou igual a 1.")
else:
    print("Por favor, digite um número menor ou igual a 100.")
```
