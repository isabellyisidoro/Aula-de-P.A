```python
#2 Estrutura de decisões (if, elif e else)

# Escrever um programa que compara 02 números e mostre qual é o maior (ou se
# são iguais)

# Entradas
a = int (input("Digite o primeiro número inteiro: "))
b = int (input("Digite o segundo número inteiro: "))

# Processamento / saída

if a > b:
  print (f"O número {a} é maior que {b}")
elif b > a:
  print (f"O número {b} é maior que {a}")
else:
  print (f"Os números {a} e {b} são iguais")
```

```python
# 6 Positivo, negativo ou zero Peça um número e diga se ele é positivo, negativo ou zero

numero = float(input("digite um numero: "))

if numero > 0:
  print(f"o numero {numero} é positivo.")

elif numero < 0:
  print(f"o numero {numero} é negativo.")

else:
  print(f"o numero {numero} é negativo.")
```

```python
# 7 Calculadora simples

# Entrada 2 valores
# Processamento calcular usando peradores aritiméticos
# saída Mostrar o resultado

#entrada

numero1 = int (input("digite um numero"))
numero2 = int (input("digite o segundo numero"))

# Processamento

soma = numero1 + numero2
subtração = numero1 - numero2
multiplicação = numero1 * numero2
divisão = numero1 / numero2
div_inteira = numero1 // numero2
resto = numero1 % numero2
potencia = numero1 ** numero2

#Saída

print (f"o resultado da soma de {numero1} e {numero2} é igual: {soma}")
print (f"o resultado da subtração de {numero1} e {numero2} é igual: {subtração}")
print (f"o resultado da multiplicação de {numero1} e {numero2}  é igual: {multiplicação}")
print (f"o resultado da divisão de {numero1} e {numero2} é igual: {divisão}")
print (f"o resultado da divisão inteira de {numero1} e {numero2} é igual: {div_inteira}")
print (f"o resto da divisão de {numero1} e {numero2} é igual: {resto}")
print (f"o resultado da potencia de {numero1} e {numero2} é igual: {potencia}")
```

```python
#9 Solicite um ano e informe se ele é bissexto ou não.

# Entrada:
ano = int(input("Digite um ano: "))

# Definição da função de processamento:
def eh_bissexto(ano):
    # Regra: Divisível por 400, ou divisível por 4 E não por 100
    return (ano % 400 == 0) or (ano % 4 == 0 and ano % 100 != 0)

# Saída:
if eh_bissexto(ano):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")

# Exemplos de teste (removidos para evitar redundância com a entrada do usuário):
# print(eh_bissexto(2024)) # True
# print(eh_bissexto(1900)) # False
```

```python
#10Escrever um programa que leia o valor da compra e aplique:

# Até R$ 100 - sem desconto
# R$ 101 - a R$ 500 - 10% de desconto
#Acima de R$ 500 - desconto de 20%

#Mostrar valor a pagar depois de aplicado o desconto.

compra = float(input("Digite o valor total da compra: R$ "))

if compra > 100:
    desconto = compra - (compra * 0.10)
    print (f"o valor a pagar é {desconto}")

elif compra > 500:
    desconto = compra - (compra * 0.20)
    print (f"o valor a pagar é: {desconto}")

else:
   print  (f"você não obteve desconto, valor a pagar : {compra}")
```
